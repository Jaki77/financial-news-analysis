### Import necessary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load Dataset

```python
results = pd.read_csv("../data/task3_results.csv")
print("TASK 3 FINAL CORRELATION RESULTS")
print(results.to_string(index=False))
```

### Visualisation

```python
if 'plt' not in globals():
    import matplotlib.pyplot as plt

plt.figure(figsize=(12, 7))
bars = plt.bar(results['Ticker'], results['Pearson Corr'],
               color=['#00C853' if x > 0 else '#D50000' for x in results['Pearson Corr']])

plt.axhline(0, color='black', linewidth=1)
plt.title("News Sentiment vs Next-Day Returns â€“ Pearson Correlation", fontsize=16, pad=20)
plt.ylabel("Correlation Coefficient", fontsize=12)
plt.xlabel("Stock Ticker", fontsize=12)

# Add value labels on bars
for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + (0.01 if height > 0 else -0.02),
             f"{height:.3f}", ha='center', va='bottom' if height > 0 else 'top',
             fontweight='bold', fontsize=11)

plt.grid(axis='y', alpha=0.3)
plt.ylim(-0.3, 0.4)
plt.tight_layout()
plt.show()
```
